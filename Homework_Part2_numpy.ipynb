# 3- Write a NumPy code line(s) to test whether any of the elements of an input array is non-zero.

# 4- Write a NumPy code line(s) to compute the x and y coordinates for points on a sine curve and plot the points using matplotlib.

```python
x = np.linspace( 0, 2*pi, 100 )
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

# 5- Write a NumPy code line(s) to extract all numbers which are less and greater than aspecified integer in an input array.

```python
arr2 = np.array([1,2,0,0,3,0,6,10,5,2,3,7])
choice = 3
print("The numbers that is less than the choosen number = ", arr2[arr2 < choice])
print("The numbers that is greater than the choosen number = ", arr2[arr2 > choice])
```
